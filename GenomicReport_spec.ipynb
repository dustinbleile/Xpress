### Group and Filter Results

Categories are decided by analysts:
 * Approved therapies for this cancer
 * Approved in other cancer
 * Other therapies
 * Prognostic
 * Diagonist
 * Uncharacterized - (Related interest)
 * Novel - (For manual review)
 * (Excluded - post mortem - why missed)

Categories are Ranked and mutually exclusive.

Knowledgebase matches table shows these results showing (mostly) standard data.
Oraganized by:
 * Header of event (eg. SV std rep.) with gene-level details (copy number - std rep) and sample source.
 * Matched KB event details in dropdown.

Detailed tables - Subcategorize by type of event (Small mut, SV, expression, copy number)
 * These are actually GSC (experimental parameter) specific details about events above.
 * Includes a few custom analysis & combinations:
  * SV expressed w/ DNA support
  * combination - lowly expressed TSG with copy loss.
