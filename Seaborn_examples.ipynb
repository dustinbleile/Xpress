```python
import os, sys, glob
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Use seaborn style
sns.set()
```

```python
# load example data

tips = sns.load_dataset("tips")
print('tips')
print(tips.head())

titanic = sns.load_dataset("titanic")
print('\ntitanic')
print(titanic.head())

print('\niris')
iris = sns.load_dataset("iris")
print(iris.head())
```

```python
g = sns.FacetGrid(titanic,
                  col="survived",
                  row="sex")
g = g.map(plt.hist,"age")
```

```python
#sns.factorplot # - outdated
sns.factorplot(x="pclass",
            y="survived",
            hue="sex",
            data=titanic)
sns.catplot(x="pclass",
            y="survived",
            hue="sex",
            data=titanic,
            kind='strip',
           )
```

```python
 sns.lmplot(x="sepal_width", #Plot data and regression model fits
            y="sepal_length", #across a FacetGrid
            hue="species",
            data=iris)
```

```python
h = sns.PairGrid(iris)
h = h.map(plt.scatter)

sns.pairplot(iris)
```

```python
i = sns.JointGrid(x="sepal_length", #"x", # Grid for bivariate plot with marginal
                  y="sepal_width", #y", #univariate plots
                  data=iris)
i = i.plot(sns.regplot,
           sns.distplot)

sns.jointplot("sepal_length", # Plot bivariate distribution
              "sepal_width",
              data=iris,
              kind='kde')
```

```python
sns.stripplot(x="species", # Scatterplot with one
              y="petal_length",  # categorical variable
              data=iris)
```

```python
sns.swarmplot(x="species", #Categorical scatterplot with
              y="petal_length", #non-overlapping points
              data=iris)
```

```python
sns.barplot(x="sex", # Show point estimates and
 y="survived", # confidence intervals with
 hue="class",  # scatterplot glyphs
 data=titanic)
```

```python
sns.countplot(x="deck", # Show count of observations
              data=titanic,
              palette="Greens_d")
```

```python
sns.pointplot(x="class", # Show point estimates and
              y="survived", # confidence intervals as
              hue="sex", # rectangular bars
              data=titanic,
              palette={"male":"g", "female":"m"},
              markers=["^","o"],
              linestyles=["-","--"])
```

```python
sns.boxplot(x="alive", # Boxplot
            y="age",
            hue="adult_male",
            data=titanic)
```

```python
sns.boxplot(data=iris, orient="h")
```

```python
sns.violinplot(x="age", # Violin plot
               y="sex",
               hue="survived",
               data=titanic)
```

```python
f, ax = plt.subplots(figsize=(5,7)) 
sns.regplot(x="sepal_width",  # Plot data and a linear regression
             y="sepal_length",  # model fit
             data=iris,
             ax=ax,
            )
```

```python
plot = sns.distplot(iris['sepal_width'], # Plot univariate distribution
                    kde=True,
                    rug=True,
                    color="b")
```
